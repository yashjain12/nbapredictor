```python
#Hi, I'll be showing this to predict NBA games. Our training dataset uses games from 2015 to 2022.
import pandas as pd
df = pd.read_csv("/kaggle/input/nbagameinfo/nba_games (1).csv", index_col=0)
df
```

```python
df = df.sort_values("date")
df
```

```python
df = df.reset_index(drop=True)

df.drop(["mp.1","mp_opp.1", "index_opp"],axis = 1,inplace = True)
```

```python
def add_target(team):
    team["target"] = team["won"].shift(-1)
    return team

#Add a "target" or "predictions" column that has values of if the game
#that was played previously was won or lost
df = df.groupby("team",group_keys=False).apply(add_target)

#Example for Los Angeles Lakers
df[df["team"] == "LAL"]
```

```python
#Assign Non-Null Values in Target to 2
df["target"][pd.isnull(df["target"])] = 2
df["target"] = df["target"].astype(int, errors="ignore")

nulls = pd.isnull(df)
nulls = nulls.sum()
nulls = nulls[nulls > 0]

valid_columns = df.columns[~df.columns.isin(nulls.index)]
df = df[valid_columns].copy()
df
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import RidgeClassifier
from sklearn.linear_model import LogisticRegressionCV
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

#Use Ridge Regression to weight features
rr = RidgeClassifier(alpha = 0.05)

#Using a Time Series to create the splits, which means that our future predictions consider historical data
split = TimeSeriesSplit(n_splits = 8)

sfs = SequentialFeatureSelector(rr, n_features_to_select = 20, direction = "forward", cv = split)

#Remove Non-Numerical Columns
removed_columns = ["season", "date", "won", "target", "team", "team_opp"]

selected_columns = df.columns[~df.columns.isin(removed_columns)]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

#Scale the columns so all values fall between 0 and 1. This also makes it better for Ridge Regression
df[selected_columns] = scaler.fit_transform(df[selected_columns])
```

```python
sfs.fit(df[selected_columns], df["target"])
```

```python
#These are the list of predictors that are deemed important after fitting

predictors = list(selected_columns[sfs.get_support()])
print(predictors)
```

```python
#We'll define this function to predict the outcomes for the current season
#given information from previous seasons.
#Then we append these results onto 1 dataframe.

def backtest(data,model, predictors, start = 2):
    all_predictions = [    ]
    seasons = [2016,2017,2018,2019,2020,2021,2022]
    for i in range(start,len(seasons)):
        train = data[data["season"] < seasons[i]]
        test = data[data["season"] == seasons[i]]
        model.fit(train[predictors], train["target"])
        preds =  pd.Series(model.predict(test[predictors]), index = test.index)
        combined = pd.concat([test["target"], preds], axis = 1)
        combined.columns = ["actual", "prediction"]
        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
#Testing process of our first model
predictions = backtest(df,rr,predictors)

correct = 0
for ind in predictions[predictions["actual"] != 2].index:
    if(predictions['actual'][ind] == predictions['prediction'][ind]):
        correct = correct + 1

#Print the percentage of games that our model correctly predicted
#We see that this result is only 54.5%, which is considerably worse
#than the accuracy when predicting the winner by comparing winning percentage at the time of the match
print(correct/predictions[predictions["actual"] != 2].shape[0])
```

```python
#To create a better model, we have to stress multiple factors even more:
#home/away game
#team record (since NBA teams roster/strength change quite frequently, we'll use the winning percentage in last 25 games)
#Using this improved model, we can see if there are any changes to the accuracy
def find_team_averages(team):
    return team.rolling(25).mean()

def shift_col(team, col_name):
    next_col = team[col_name].shift(-1)
    return next_col

def add_col(df, col_name):
    return df.groupby("team", group_keys = False).apply(lambda x:shift_col(x, col_name))

df_rolling = df_rolling.groupby(["team","season"], group_keys = False).apply(find_team_averages)

rolling_cols = [f"{col}_new" for col in df_rolling.columns]

df_rolling.columns = rolling_cols

df = (pd.concat([df, df_rolling], axis = 1)).dropna()

df["home_next"] = add_col(df, "home")
df["team_opp_next"] = add_col(df, "team_opp")
df["date_next"] = add_col(df, "date")
full = df.merge(df[rolling_cols + ["team_opp_next","date_next", "team"]],left_on=["team", "date_next"],right_on = ["team_opp_next", "date_next"])
removed_columns = list(full.columns[full.dtypes == "object"]) + removed_columns

selected_columns = full.columns[~full.columns.isin(removed_columns)]
sfs.fit(full[selected_columns], full["target"])
predictors = list(selected_columns[sfs.get_support()])
predictions = backtest(full, rr, predictors)

accuracy_score(predictions["actual"], predictions["prediction"])
#Our Accuracy results in 65.2%, a much more improved result than before.
```
